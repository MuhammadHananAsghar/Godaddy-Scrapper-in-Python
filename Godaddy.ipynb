```python
import requests
from bs4 import BeautifulSoup as Soup
from fake_useragent import UserAgent
ua = UserAgent()
import pandas as pd
```

```python
def GoddayDomain():
    search = input("Enter Domain: ")
    html = requests.get(f"https://pk.godaddy.com/domainfind/v1/search/spins?&q={str(search)}&pagestart=0&pagesize=60",headers={"User-Agent":f"{ua.random}"}).json()
    product_ids = []
    for i in range(len(html['Products'])):
        val = html['Products'][i]['ProductId']
        product_ids.append(val)
    product_prices = []
    for i in range(len(html['Products'])):
        val = html['Products'][i]['PriceInfo']['CurrentPriceDisplay']
        product_prices.append(val)
    product_tlds = []
    for i in range(len(html['Products'])):
        val = html['Products'][i]['Tld']
        product_tlds.append(val)
    data = {
        "TLD":product_tlds,
        "Price":product_prices,
        "IDS":product_ids
    }
    prod_ids = []
    for i in range(len(html['RecommendedDomains'])):
        val = html['RecommendedDomains'][i]['ProductId']
        prod_ids.append(val)
    prod_domain = []
    for i in range(len(html['RecommendedDomains'])):
        val = html['RecommendedDomains'][i]['Fqdn']
        prod_domain.append(val)
    data2 = {
        "IDS":prod_ids,
        "Domain Name":prod_domain
    }
    df1 = pd.DataFrame(data)
    df2 = pd.DataFrame(data2)
    dataFrame = pd.merge(df1,df2,how="inner")
    dataFrame = dataFrame.sort_values('Domain Name')
    dataFrame = dataFrame['Provider'] = 'Godaddy'
    return dataFrame
```

```python
df = GoddayDomain()
```

```python
df['Provider'] = 'Godaddy'
```
